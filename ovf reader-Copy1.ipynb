```python
%matplotlib widget
from NetSocks import NSClient
import numpy as np
import os
import matplotlib.pyplot as plt

ns = NSClient()
```

```python
SaveFolder_directory = "/home/ashwin/Documents/Spin_wave_pbc/Images" 
if not os.path.exists(SaveFolder_directory):
    os.makedirs(SaveFolder_directory)


###############################
##### Specify run number ######
###############################
run = 1


SaveImage_directory = SaveFolder_directory + f"/Run_{run}"
if not os.path.exists(SaveImage_directory):
    os.makedirs(SaveImage_directory)
```

```python
import glob

file_names = glob.glob("/media/ashwin/TOSHIBA EXT/Spin-waves/ovf-files/*.ovf", recursive=True)
#Pretty bad at list comphrehension methods
#file_nums="".join(c for c in i if c.isdigit() for i in file_names)
file_nums = []
for i in file_names:
    s=""
    for char in i:
        if char.isdigit():
            s+=char
    file_nums.append(int(s))

# file_list = np.sort(np.array(files))
# print(file_list)
sort = np.sort(file_nums) #It's already in a numpy array haha
#print(sort)

file_list = np.array(["/media/ashwin/TOSHIBA EXT/Spin-Waves/ovf-files/mag_val_at_"+str(i)+"fs.ovf" for i in sort]) #two for loops not the prettiest code 
```

## Plotting

```python
dummy = 1
for i in range(len(file_list[:])):

    
    vec, n, rect = ns.Read_OVF2(str(file_list[i]))
    #vec, n, rect = ns.Read_OVF2("/media/ashwin/TOSHIBA EXT/Spin-waves/ovf-files/mag_val_at_1fs.ovf")
    x = vec[:256*256, 0]
    y = vec[:256*256, 1]
    z = vec[:256*256, 2]

    m = np.sqrt(x**2 + y**2 + z**2)
    m = m.reshape(256,256)

    z = z.reshape(256,256)
    #print(np.mean(m))

    plt.suptitle("time = "+str(sort[i])+" fs") 
    plt.figure(figsize = (10,10))
    #plt.imshow(m, cmap = 'inferno')
    plt.imshow(z, cmap = 'inferno')
    #print(i, '\n', m.shape)

    
    plt.savefig(f"{SaveImage_directory}/{dummy}.png")

    #plt.show()
    #plt.close()
    dummy+=1
```
